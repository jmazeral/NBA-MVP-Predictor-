```python
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
from io import StringIO
```

```python
# years we are scraping from
years = list(range(1989, 2025))

# webscraping url
url = f"https://www.basketball-reference.com/awards/awards_YEAR.html"
```

```python
# saving html data to mvps folder to be used later
for year in years:
    url = f"https://www.basketball-reference.com/awards/awards_{year}.html"
    data = requests.get(url)
    with open("mvps/{}.html".format(year), "w+", encoding="utf-8") as file:
        file.write(data.text)
    # basketballreference hates me so i am sleeping so i dont get banned again
    time.sleep(10)
```

```python
# taking our scraped files and getting only the mvp charts
mvpdfs = []
for year in years:
    with open("mvps/{}.html".format(year), encoding="utf-8") as file:
        page = file.read()
    soup = BeautifulSoup(page, "html.parser")
    # useless header at the top of the chart
    soup.find("tr", class_="over_header").decompose()
    mvp_chart = soup.find(id="mvp")
    df = pd.read_html(StringIO(str(mvp_chart)))[0]
    df["Year"] = year
    mvpdfs.append(df)
mvps = pd.concat(mvpdfs)
mvps.head()
```
